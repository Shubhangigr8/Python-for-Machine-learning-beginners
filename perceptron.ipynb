```python
import numpy as np
class preceptron(object):
    def __init__(self,eta=0.01,n_iter=20):
        self.eta=eta #learning rate(between 0.0 and 1.0)
        self.n_iter=n_iter
    def fit(self,x,y):
        self.w_=np.zeros(1+x.shape[1])
        k=1
        for _ in range(self.n_iter):
            print("iteration",k)
            flag=0
            for xi,target in zip(x,y):
                print("input is",xi,"target is",target)
                print('Output is ',self.predict(xi))
                error = target - self.predict(xi)
                print("actual weights",self.w_)
                if error!=0:
                    flag=1
                    update=self.eta*target
                    self.w_[1:]+=update*xi
                    self.w_[0]+=update
                    print("update weight is", self.w_,"\n")
                k+=1
            if flag == 0:
                break
        return self
    def net_input(self,x):
        #"""calculate net input"""
        y=np.dot(x,self.w_[1:])+self.w_[0]
        #print(y)
        return y
    def predict(self,x):
        return np.where(self.net_input(x)>0.0,1,-1)               
                            #input
                       
```

```python
 x=np.array([[1,1],[-1,1],[1,-1],[-1,1]])
print('inputs:',x)
                        #target
y=np.array([1,-1,-1,1])
                        ##call perceptron
ppn=preceptron(eta=1,n_iter=10)
ppn.fit(x,y)
print("predict:",ppn.predict(x))
```
