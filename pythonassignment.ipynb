```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(-10,10,100)
y=x
plt.plot(x,y,'r--')
plt.title("Identity")
plt.show()
```

```python
x=np.linspace(-10,10,100)
y=1/(1+np.exp(-x))
plt.plot(x,y,'r--')
plt.title("Sigmoidal")
plt.show()
```

```python
x=np.linspace(-10,10,100)
y=(1-np.exp(-x))/(1+np.exp(-x))
plt.plot(x,y,'r--')
plt.title("Polar")
plt.show()
```

```python
x=np.linspace(-10,10,100)
y=-x
plt.plot(x,y,'r--')
plt.title("Bipolar")
plt.show()
```
